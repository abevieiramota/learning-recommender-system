```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
user_movie = pd.DataFrame([\
                             ('Abelardo', 'Life of Pi'),
                             ('Abelardo', 'Donnie Darko'),
                             ('Abelardo', 'Titanic'),
                             ('Cris', 'Chuck'),
                             ('Cris', 'Apocalypse now'),
                             ('Obama', 'Bastardos Inglórios'),
                             ('Obama', 'Titanic'),
                             ('Obama', 'Brasileirinhas'),
                             ('ElTaradão', 'Francesinhas'),
                             ('ElTaradão', 'Espanholinhas')
                             ], columns=['user', 'movie'])
```

```python
movie_tag_ = pd.DataFrame([\
                         ('Life of Pi', 'Fantasia'),
                         ('Life of Pi', 'Surto'),
                         ('Life of Pi', 'Oscar'),
                         ('Life of Pi', 'Drama'),
                         ('Life of Pi', 'Aventura'),
                         ('Donnie Darko', 'Surto'),
                         ('Donnie Darko', 'Drama'),
                         ('Donnie Darko', 'Meio maluco'),
                         ('Titanic', 'Drama'),
                         ('Titanic', 'Romance'),
                         ('Titanic', 'Leonardo DiCaprio'),
                         ('Chuck', 'Terror'),
                         ('Chuck', 'Comédia'),
                         ('Apocalypse now', 'Drama'),
                         ('Apocalypse now', 'Guerra'),
                         ('Bastardos Inglórios', 'Guerra'),
                         ('Bastardos Inglórios', 'Drama'),
                         ('Bastardos Inglórios', 'Aventura'),
                         ('Brasileirinhas', 'Pornô'),
                         ('Brasileirinhas', 'Nacional'),
                         ('Brasileirinhas', 'Milf'),
                         ('Francesinhas', 'Pornô'),
                         ('Francesinhas', 'Internacional'),
                         ('Espanholinhas', 'Pornô'),
                         ('Espanholinhas', 'Internacional'),
                         ('Espanholinhas', 'Milf')], columns=['movie', 'tag'])
```

# User x tag

```python
user_tag = pd.merge(user_movie, movie_tag_).groupby(['user', 'tag']).count().unstack().fillna(0)
user_tag
```

```python
user_tag = user_tag.div(user_tag.sum(axis=1), axis=0)
user_tag
```

# Movie x tag

```python
movie_tag = pd.get_dummies(movie_tag_, columns=['tag'], prefix='', prefix_sep='').groupby('movie').sum()
movie_tag
```

```python
movie_tag = movie_tag.div(movie_tag.sum(axis=1), axis=0)
movie_tag
```

# User x Movie

```python
um = np.dot(user_tag.values, movie_tag.values.T)
um
```

### removendo filmes já vistos

```python
usu_ja_visto = pd.get_dummies(user_movie, columns=['movie'], prefix='', prefix_sep='').groupby('user').sum().applymap(lambda x: {0.: 1., 1.:0.}[x])
```

```python
um = um * usu_ja_visto.values
```

```python
um = pd.DataFrame(um, columns=movie_tag.T.columns, index=user_tag.index)
```
